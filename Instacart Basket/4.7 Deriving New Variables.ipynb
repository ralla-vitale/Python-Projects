### I. If statement

```python
# i. defining price_label with if statement

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying price_label function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# frequency

df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

#### Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# 1. frequency

df_orders_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# 2. empty list and for loop

busyness = []

for value in df_orders_prods_merged['orders_day_of_week']:
  if value == 0:
    busyness.append('Busiest days')
  elif value == 4:
    busyness.append('Slowest days')
  else:
    busyness.append('Regularly day')
```

```python
# 3. adding busiest day and orders_prods_merged dataset into busyness column

df_orders_prods_merged['Busiest_days'] = busyness
```

```python
# 4. frequency column

df_orders_prods_merged['Busiest_days'].value_counts(dropna = False)
```
